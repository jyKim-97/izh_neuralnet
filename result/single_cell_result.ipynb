```python
import numpy as np
import matplotlib.pyplot as plt
import nnpy.Izhtools as it
import nnpy.visu_tools as visu
```

```python
obj_rs = it.Reader("../data/one_rs"); obj_rs.read_uv()
obj_fs = it.Reader("../data/one_fs"); obj_fs.read_uv()
obj_ib = it.Reader("../data/one_ib"); obj_ib.read_uv()
obj_ch = it.Reader("../data/one_ch"); obj_ch.read_uv()
```

```python
plt.figure(dpi=150, figsize=(8, 3))
labels = ['RS', 'IB', 'CH']

for obj, lb in zip([obj_rs, obj_ib, obj_ch], labels):
    plt.plot(obj.ts, obj.vs, lw=1.2, label=lb)
    
plt.legend(fontsize=10, edgecolor='k')
plt.xlim([900, 1700])
plt.xlabel('time (ms)', fontsize=12)
plt.ylabel('voltage (mV)', fontsize=12)
plt.show()
```

```python
plt.figure(dpi=150, figsize=(6, 5))
labels = ['RS', 'IB', 'CH']

for n, obj in enumerate([obj_rs, obj_ib, obj_ch]):
    plt.subplot(3,4,n+1)
    plt.plot(obj.ts, obj.vs, 'k', lw=1)
    plt.xlim([900, 1300])
    plt.ylabel('%s\nvoltage (mV)'%(labels[n]), fontsize=10)
    plt.grid(True)
    plt.yticks([-70, -35, 0, 35], fontsize=8)
    plt.xticks(fontsize=8)
    
    
    
plt.xlabel('time (ms)', fontsize=10)
plt.tight_layout()
plt.show()
```
